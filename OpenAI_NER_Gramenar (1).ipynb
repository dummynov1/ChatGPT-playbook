```python
from promptify import Prompter,OpenAI, Pipeline
# Define the API key for the OpenAI model
# Create an instance of the OpenAI model, Currently supporting Openai's all model, In future adding more generative models from Hugginface and other platforms
model        = OpenAI(api_key)
prompter     = Prompter('ner.jinja')
pipe         = Pipeline(prompter , model)


from PyPDF2 import PdfReader

# Open the PDF file in read-binary mode
with open('/content/Resume - Sample.pdf', 'rb') as pdf_file:
    # Create a PdfReader object
    pdf_reader = PdfReader(pdf_file)

    # Initialize an empty string to store the PDF text
    pdf_text = ""

    # Iterate through each page in the PDF
    for page_num in range(len(pdf_reader.pages)):
        # Get the text from the current page
        page = pdf_reader.pages[page_num]
        pdf_text += page.extract_text()

# Now, 'pdf_text' contains the text content of the PDF file

print(pdf_text)
```

```python
# If want to perform NER with custom tags only (hangling out-of-bounds prediction) prompt
result       = pipe.fit(pdf_text,
                          domain  = 'resume',
                          labels  = ["Name", "Email Address", "Graduation Year", "College Name", "Degree", "Location", "Years of Experience", "Companies worked at", "Designation", "Skills", "UNKNOWN" ])
print(result)
```
