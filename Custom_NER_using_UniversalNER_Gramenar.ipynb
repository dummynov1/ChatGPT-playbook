```python
from langchain import HuggingFacePipeline
from transformers import AutoTokenizer, AutoModelForCausalLM
import transformers
import torch

model = "Universal-NER/UniNER-7B-all"

tokenizer = AutoTokenizer.from_pretrained(model)
pipeline = transformers.pipeline(
    "ner", #task
    model=model,
    tokenizer=tokenizer,
    trust_remote_code=True,
    device_map="auto",
    torch_dtype=torch.bfloat16,
    max_length=1000,
    do_sample=True,
    top_k=10,
    num_return_sequences=1,
    eos_token_id=tokenizer.eos_token_id,

)
```

```python
llm = HuggingFacePipeline(pipeline = pipeline, model_kwargs = {'temperature':0})
```

```python
from langchain import PromptTemplate,  LLMChain

template = """
              A virtual assistant answers questions from a user based on the provided text.
              USER: Text: {input_text}
              ASSISTANT: I’ve read this text.
              USER: What describes {entity_type} in the text?
              ASSISTANT:
           """

prompt = PromptTemplate(template=template, input_variables=["input_text","entity_type"])

llm_chain = LLMChain(prompt=prompt, llm=llm)
```

```python
from PyPDF2 import PdfReader

# Open the PDF file in read-binary mode
with open('/content/Resume - Sample.pdf', 'rb') as pdf_file:
    # Create a PdfReader object
    pdf_reader = PdfReader(pdf_file)

    # Initialize an empty string to store the PDF text
    pdf_text = ""

    # Iterate through each page in the PDF
    for page_num in range(len(pdf_reader.pages)):
        # Get the text from the current page
        page = pdf_reader.pages[page_num]
        pdf_text += page.extract_text()

# Now, 'pdf_text' contains the text content of the PDF file

print(pdf_text)
```

```python
# Define entity types
entity_types = ["Name", "Email Address", "Graduation Year", "College Name", "Degree", "Location", "Years of Experience", "Companies worked at", "Designation", "Skills", "UNKNOWN" ]

# Loop through entity types and extract entities
for entity_type in entity_types:
    response = llm_chain.run({"input_text": pdf_text, "entity_type": [entity_type]})
    print(f"Entities of type {entity_type}: {response[0]['ner']}")
```
